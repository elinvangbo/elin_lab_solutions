```python
new_cols = [col.replace(" ", "_").lower() for col in data]
data.columns = new_cols
```

```python
numericals = data.select_dtypes(np.number)
```

```python
#Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
```

```python
def df_dist(df):
    sns.set(rc={'figure.figsize':(8,4)})
    for i, col in enumerate(df):
        plt.figure(i)
        sns.distplot(df[col], color = list(BASE_COLORS.keys())[i])
        plt.show()
```

```python
print(df_dist(numericals))
```

```python
def df_dist2(df):
    sns.set(rc={'figure.figsize':(8,4)})
    for i, col in enumerate(df):
        plt.figure(i)
        ax = sns.distplot(df[col], color = list(BASE_COLORS.keys())[i])
        ax.set_yscale('log')
        plt.show()
        
```

```python
print(df_dist2(numericals))
```
