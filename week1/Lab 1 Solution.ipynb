```python
def our_magic_function(x):
    x = str(x)
    if x.endswith("%"): 
        return float(x[:-1])/100
    else:
        return float(x)
```

```python
data["CUSTOMER LIFETIME VALUE"] = data["CUSTOMER LIFETIME VALUE"].apply(our_magic_function)
```
